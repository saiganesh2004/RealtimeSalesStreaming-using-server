```python
import pandas as pd
import random
import time
import requests
from datetime import datetime

# Your Apps Script Web App URL
URL = "https://script.google.com/macros/s/AKfycbzTo9FVidrzRt14h2WnCmDyQp-khDzM_FdlKLb0-W0MlMl870DqsGEkQCJpLs-kT1Jn3w/exec"

def generate_sales_data():
    return {
        "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
        "sales": random.randint(100, 1000)
    }

# Send data continuously
while True:
    data = generate_sales_data()

    response = requests.post(URL, json=data)
    print("Sent:", data, "| Response:", response.text)

    time.sleep(1)  # send every 1 seconds
```
